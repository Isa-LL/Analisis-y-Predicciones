```python
#Funcion para predecir los datos según la formula de regresión lineal

import pandas as pd


df = pd.read_csv('data_science_salaries.csv')

x = df['work_year']
y = df['salary_in_usd']

n = len(x)
sum_x = x.sum()
sum_y = y.sum()
sum_xy = (x * y).sum()
sum_x_squared = (x ** 2).sum()


m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
b = (sum_y - m * sum_x) / n


year_2025 = 2024
salary_2025 = m * year_2025 + b

print("Pendiente (m):", m)
print("Intersección (b):", b)
print("Salario estimado en 2025:", salary_2025)


m_formatted = f"{m:,.2f}"
b_formatted = f"{b:,.2f}"
salary_2025_formatted = f"{salary_2025:,.2f}"

print("#########################################")

print("Pendiente (m):", m_formatted)
print("Intersección (b):", b_formatted)
print("Salario estimado en 2025: $", salary_2025_formatted)

```

```python
#Encapsulamiento como funcion para usarla en javascript

import pandas as pd

def predict_salary(job_title, year, csv_file='data_science_salaries.csv'):
   

    df = pd.read_csv(csv_file)

    df_filtered = df[df['job_title'] == job_title]
    
    if df_filtered.empty:
        return f"No se encontraron datos para el puesto: {job_title}"
  
    x = df_filtered['work_year']
    y = df_filtered['salary_in_usd']
    
    n = len(x)
    sum_x = x.sum()
    sum_y = y.sum()
    sum_xy = (x * y).sum()
    sum_x_squared = (x ** 2).sum()

    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
    b = (sum_y - m * sum_x) / n

    
    predicted_salary = m * year + b

    return {
        "job_title": job_title,
        "year": year,
        "predicted_salary": f"{predicted_salary:,.2f}",
        "m": f"{m:,.2f}",
        "b": f"{b:,.2f}"
    }

result = predict_salary('Data Engineer', 2025)
print(result)
```

```python
#Funcion para sacar valores únicos de una columna
import pandas as pd

def get_unique_column_values(csv_file, column_name):
    try:
    
        df = pd.read_csv(csv_file)
        
        
        if column_name not in df.columns:
            return f"La columna '{column_name}' no se encuentra en el archivo CSV."

        
        unique_values = df[column_name].dropna().unique()

        return unique_values.tolist() 
    except Exception as e:
        return f"Se produjo un error: {e}"


unique_values = get_unique_column_values('data_science_salaries.csv', 'job_title')
print(unique_values)



```

```python
#Está funcion cuenta los elemenos repetidos en una columna y los guarda según los elementos en un JSON

import pandas as pd
import json

def count_repeated_values_to_json(csv_file, column_name, output_json_file):
    try:
        
        df = pd.read_csv(csv_file)
        
        
        if column_name not in df.columns:
            return f"La columna '{column_name}' no se encuentra en el archivo CSV."

       
        value_counts = df[column_name].value_counts()

        
        counts_dict = value_counts.to_dict()

        
        with open(output_json_file, 'w', encoding='utf-8') as f:
            json.dump(counts_dict, f, ensure_ascii=False, indent=4)

        return f"Las cuentas se han guardado en el archivo '{output_json_file}'."
    except Exception as e:
        return f"Se produjo un error: {e}"


result = count_repeated_values_to_json('data_science_salaries.csv', 'job_title', 'job_title_counts.json')
print(result)

```

```python
#Está funcion cuenta la frecuencia de apariciones de los titulos de trabajo y los que son menores a 600 los agrupa en una
#categoría de otros para usarlo en la grafica ya que lo exportamos a un JSON

import json

def group_elements_by_frequency(input_json_file, output_json_file, threshold=600):
    try:
        
        with open(input_json_file, 'r', encoding='utf-8') as f:
            data = json.load(f)

        
        if not isinstance(data, dict):
            return "El archivo JSON debe contener un diccionario con pares clave-valor."

      
        grouped_data = {}
        otros_sum = 0

        
        for key, value in data.items():
            if value < threshold:
                otros_sum += value 
            else:
                grouped_data[key] = value 

        
        if otros_sum > 0:
            grouped_data["Otros"] = otros_sum

       
        with open(output_json_file, 'w', encoding='utf-8') as f:
            json.dump(grouped_data, f, ensure_ascii=False, indent=4)

        return f"El archivo JSON agrupado se ha guardado en '{output_json_file}'."
    except Exception as e:
        return f"Se produjo un error: {e}"


result = group_elements_by_frequency('job_title_counts.json', 'grouped_job_title_counts.json')
print(result)
```

```python
#Funcion para filtrar exportar a un JSON los valores de una columna, filtrados y convertidos a lista.

import pandas as pd
import json


def csv_column_to_json(csv_file, column_name, json_file):
    try:
    
        df = pd.read_csv(csv_file)
        
    
        if column_name not in df.columns:
            return f"La columna '{column_name}' no se encuentra en el archivo CSV."
        
    
        column_values = df[column_name].dropna().tolist()
        
      
        with open(json_file, 'w', encoding='utf-8') as f:
            json.dump(column_values, f, ensure_ascii=False, indent=4)
        
        return f"Los valores de la columna '{column_name}' se han guardado en el archivo '{json_file}'."
    except Exception as e:
        return f"Se produjo un error: {e}"


result = csv_column_to_json('data_science_salaries.csv', 'job_title', 'job_titles.json')
print(result)
```

```python
#Funcion para contar los valores de una la columna employed y fitrar los 4 primeros más repetidos y exportarlos a un JSON

import pandas as pd
import json

def count_top_countries(csv_file, column_name='employee_residence', top_n=4, output_file='country_counts.json'):
    
    df = pd.read_csv(csv_file)

    
    country_counts = df[column_name].value_counts()

    
    top_countries = country_counts.head(top_n)
    others_count = country_counts.iloc[top_n:].sum()

   
    result = {country: int(count) for country, count in top_countries.items()}
    result["Otros"] = int(others_count)

   
    with open(output_file, 'w') as f:
        json.dump(result, f, indent=4)

    return result

csv_file = 'data_science_salaries.csv'
output = count_top_countries(csv_file)
print(json.dumps(output, indent=4))

```

```python
#Calcula el promedio del salario de los año de los trabajos principales

import pandas as pd

def predict_salary(job_title, year, csv_file='data_science_salaries.csv'):
    df = pd.read_csv(csv_file)

    df_filtered = df[df['job_title'] == job_title]
    if df_filtered.empty:
        return f"No se encontraron datos para el puesto: {job_title}"
  
    x = df_filtered['work_year']
    y = df_filtered['salary_in_usd']
    
    n = len(x)
    sum_x = x.sum()
    sum_y = y.sum()
    sum_xy = (x * y).sum()
    sum_x_squared = (x ** 2).sum()

    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
    b = (sum_y - m * sum_x) / n

    predicted_salary = m * year + b

    return {
        "job_title": job_title,
        "year": year,
        "predicted_salary": f"{predicted_salary:,.2f}"
    }


titles = ["Data Engineer", "Data Scientist", "Data Analyst", "Machine Learning Engineer"]
years = [2020, 2021, 2022, 2023, 2024, 2025]


results = []
for title in titles:
    for year in years:
        result = predict_salary(title, year)
        results.append(result)

for res in results:
    print(res)
```

```python
#Calcula el promedio de los otros trabajos excluyendo a los principales

import pandas as pd
import json

def calculate_others_average(csv_file, titles_to_exclude, years, output_file='others_average.json'):
    
    df = pd.read_csv(csv_file)
    
   
    df_others = df[~df['job_title'].isin(titles_to_exclude)]
    
    if df_others.empty:
        return f"No se encontraron trabajos diferentes a {titles_to_exclude}"
    
  
    df_others = df_others[df_others['work_year'].isin(years)]
    
   
    averages = df_others.groupby('work_year')['salary_in_usd'].mean().to_dict()
    
   
    result = {
        "excluded_titles": titles_to_exclude,
        "average_salaries_for_others": {int(year): f"{avg:,.2f}" for year, avg in averages.items()}
    }
    
    
    with open(output_file, 'w') as f:
        json.dump(result, f, indent=4)
    
    return result


titles = ["Data Engineer", "Data Scientist", "Data Analyst", "Machine Learning Engineer"]
years = [2020, 2021, 2022, 2023, 2024, 2025]


csv_file = 'data_science_salaries.csv'
output = calculate_others_average(csv_file, titles, years)


print(json.dumps(output, indent=4))

```
